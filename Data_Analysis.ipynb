```python
data_dir_path = "/Users/chiang/数据/新浪微博新闻分类/cnews/"

vocab_data_path = data_dir_path + "cnews.vocab.txt"

train_data_path = data_dir_path + "cnews.train.txt"
val_data_path = data_dir_path + "cnews.val.txt"
test_data_path = data_dir_path + "cnews.test.txt"
```

```python
def read_data(file_path):
    with open(file_path) as file:
        result = []
        line = file.readline()
        result.append(line)
        while line:
            line = file.readline()
            result.append(line)
    return result
```

```python
result = read_data(train_data_path)
```

```python
def reshape_data(data_arr):
    labels = []
    texts = []
    for idx in range(len(data_arr)):
        temp = data_arr[idx].strip("\n").split("\t")
        if len(temp) == 2:
            labels.append(temp[0])
            texts.append(temp[1])
        else:
            print("有问题的文本:{}".format(data_arr[idx]))
    return labels, texts
labels, texts = reshape_data(result)
print("训练集情况：")
print("共有{}条样本".format(len(texts)))
print("共有{}个标签".format(len(labels)))
```

```python
labels_counter = dict()
for idx in range(len(labels)):
    if labels[idx] not in labels_counter:
        labels_counter[labels[idx]] = 1
    else:
        labels_counter[labels[idx]] += 1
print(labels_counter)
```

```python
length_counter = 0
for idx in range(len(texts)):
    length_counter += len(texts[idx])
print("text平均长度为：{}".format(length_counter / len(texts)))
```
